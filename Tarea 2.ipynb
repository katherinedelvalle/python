```python
import os
import pandas as pd
```

Accediendo a los datos de la felicidad

```python
url = 'https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv'

# Cargar el archivo CSV en un DataFrame
datos_felicidad = pd.read_csv(url)
datos_felicidad


```

Cálculo de la medida de tendencia central y dispersión a nivel global para el primer y último año incluido en el set de datos 

```python
cal_media = datos_felicidad.groupby('pais')['afecto_positivo'].mean()
cal_media
```

Accediendo a la información de la felicidad del primer y último año para los dos países de su elección

```python
pais_e_1 = datos_felicidad[datos_felicidad['pais'] == 'Algeria']
pais_e_1
```

```python
pais_e_2 = datos_felicidad[datos_felicidad['pais'] == 'Moldova']
pais_e_2
```

```python
max_anio_1 = pais_e_1['anio'].max()
pais_e_1['max_anio_1'] = max_anio_1
pais_e_1
```

Maximo año de Algeria

```python
maximo_anio = pais_e_1['anio'].max()
fila_maximo_anio = pais_e_1[pais_e_1['anio'] == maximo_anio]
fila_maximo_anio

```

Minimo año de Algeria

```python
min_anio = pais_e_1['anio'].min()
fila_min_anio = pais_e_1[pais_e_1['anio'] == min_anio]
fila_min_anio
```

Maximo año de Moldova

```python
maximo_anio_2 = pais_e_2['anio'].max()
fila_maximo_anio_2 = pais_e_2[pais_e_2['anio'] == maximo_anio_2]
fila_maximo_anio_2
```

Minimo año de Moldova

```python
min_anio_2 = pais_e_2['anio'].min()
fila_min_anio_2 = pais_e_2[pais_e_2['anio'] == min_anio_2]
fila_min_anio_2
```
